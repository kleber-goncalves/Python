# Analise de uso de IA nas escolas

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo o arquivo CSV
df = pd.read_csv("data/students_ai_usage.csv")
```

```python
plt.plot(df["age"], color='magenta')


#  FUNDO PRETO
ax = plt.gca()

 # fundo do grfico
plt.gca().set_facecolor('black')     
plt.gcf().set_facecolor('black')  


plt.title("Idade dos Estudantes", color='white')
plt.xlabel("Alunos", color='white', fontsize=12)
plt.ylabel("Idade", color='white', fontsize=12)


# NMEROS DOS EIXOS
plt.tick_params(axis='both', colors='white')

# BORDAS (O QUADRADO DO GRFICO)
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')


plt.grid(True)
plt.show()
```

# Pergunta : A idade interfere nas horas de estudo dirias dos estudantes?

```python



plt.figure(figsize=(8, 5))

#  FUNDO PRETO
ax = plt.gca()

 # fundo do grfico
plt.gca().set_facecolor('black')     
plt.gcf().set_facecolor('black')  


plt.scatter(df["age"],df["study_hours_per_day"],alpha=0.7, color="green")

plt.title("Idade x Horas de Estudo por Dia", color='white')
plt.xlabel("Idade", color='white', fontsize=12)
plt.ylabel("Horas de Estudo por Dia", color='white', fontsize=12)


# NMEROS DOS EIXOS 
plt.tick_params(axis='both', colors='white')

# BORDAS (O QUADRADO DO GRFICO)
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')

plt.grid(True)

plt.show()
```

- O grfico demonstra a relao entre idade e horas de estudo dirias, facilitando a identificao de padres de comportamento entre os estudantes.

# Pergunta : Estudantes de nveis de escolaridade mais altos utilizam mais IA?

```python
# Contar quantas pessoas usam IA por nvel de escolaridade
dados = df.groupby("education_level")["uses_ai"].count()




plt.bar(dados.index, dados.values, color="salmon")

plt.title("Uso da IA por Nvel de Escolaridade")
plt.xlabel("Nvel de Escolaridade" )
plt.ylabel("Quantidade")




plt.grid(True)
plt.show()
```

- O grafico demonstra a comparao entre nvel de educao e uso de IA, e permite identificar se h uma tendncia de maior uso de IA entre estudantes com nveis educacionais mais elevados.

# Pergunta : As notas melhoraram aps a introduo da IA?

```python
antes = df["grades_before_ai"].mean()
depois = df["grades_after_ai"].mean()

#  FUNDO PRETO
ax = plt.gca()

 # fundo do grfico
plt.gca().set_facecolor('black')     
plt.gcf().set_facecolor('black')  



plt.bar(["Notas antes da IA", "Notas depois da IA"], [antes, depois])

plt.title("Media de Notas antes e depois da IA", color='white')
plt.xlabel("Notas", color='white', fontsize=12)
plt.ylabel("Uso da IA", color='white', fontsize=12)


# NMEROS DOS EIXOS 
plt.tick_params(axis='both', colors='white')

# BORDAS (O QUADRADO DO GRFICO)
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')


plt.grid(True)
plt.show()
```

- O grafico demonstra a distribuio das notas dos estudantes antes e depois da introduo da IA, facilitando a comparao e anlise de possveis melhorias no desempenho acadmico.

# Pergunta : A maioria dos estudantes utiliza IA?

```python

# Contar valores da coluna uses_ai
contagem = df["uses_ai"].value_counts()

# Criar grfico de pizza
# percentuais diretamente nas fatias do grfico.
plt.pie(contagem.values,labels=contagem.index,autopct="%1.1f%%")

plt.title("Uso de IA pelos Estudantes")
plt.show()
```

- O grafico demonstra se a maioria dos estudantes utiliza IA, permitindo uma anlise rpida da adoo de tecnologias de IA entre os estudantes.
